```python
final_df=pd.merge(df,df_country,on='Country Code',how='left')
```

```python
final_df
```

```python
final_df.head(2)
```

```python
final_df.Country.unique()
```

```python
final_df.Country.value_counts()
```

```python
country_names=final_df.Country.value_counts().index
```

```python
country_values=final_df.Country.value_counts().values
country_values
```

```python
plt.pie(country_values,labels=country_names)
```

```python
plt.pie(country_values[:3],labels=country_names[:3])
```

```python
plt.pie(country_values[:3],labels=country_names[:3],autopct='%1.3f%%')
```

```python
final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text'])
```

```python
final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size()
```

```python
final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index()
```

```python
ratings=final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
ratings
```

```python
### Observation: whenever the rating is from 4.5 to 4.9 -> excellent
###              whenever the rating is from 4.0 to 4.4 -> Very good
###              whenever the rating is from 3.5 to 3.0 ->  good
###              whenever the rating is from 2.5 to 2.9 -> Average
###              whenever the rating is from 1.8 to 2.4 ->  poor
```

```python
ratings.head(2)
```

```python
sns.barplot(x='Aggregate rating',y='Rating Count',data=ratings)
```

```python
sns.barplot(x='Aggregate rating',y='Rating Count',hue='Rating color',data=ratings)
```

```python
p=['Blue','Red','Orange','Yellow','Green','Green']
```

```python
sns.barplot(x='Aggregate rating',y='Rating Count',hue='Rating color',data=ratings,palette=p)
```

```python
# Observation: Not rated count is very high
# Max no of ratings are between 2.6 to 4.6
```

```python
sns.countplot(x='Rating color',data=ratings,palette=p)
```
